```python
# Importando a biblioteca

import pandas as pd
```

> Entendendo as estruturas *Series* e *DataFrame* na prática...

```python
# Criando uma Series

clientes = ['Braund, Mr. Owen Harris',
            'Allen, Mr. William Henry',
            'Bonnell, Miss. Elizabeth']

indices = [0,1,2]
```

```python
serie = pd.Series(data=clientes,
                  index=indices,
                  name="Nomes dos Clientes")
```

```python
serie
```

- **Observação**

`Estrutura do tipo Lista: expressa através de uma sequência mutável cujos itens são separados por vírgula e delimitados por colchetes.`

Para saber mais sobre *Lists* acesse: https://docs.python.org/3/library/stdtypes.html#list

```python
# Criando um DataFrame

df = pd.DataFrame({
    "Name": ["Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",],
    "Age": [22, 35, 58],
    "Sex": ["male", "male", "female"]    
}
)
```

- **Observação**

`Estrutura do tipo Dicionário: tipo de conjunto onde os valores são indexados por chave (em geral strings ou integer). Os elementos contidos em um dicionário são pares de chave-valor`.

Para saber mais sobre *Dictionaries* acesse: https://docs.python.org/3/tutorial/datastructures.html#tut-dictionaries

```python
# Visualizando o DataFrame criado

print(df)
```

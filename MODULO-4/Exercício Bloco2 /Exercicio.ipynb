```python
def amostra_sistematica(df, step):
    indice = np.arange(1,len(df), step = step)
    amostra_sistematica = df.iloc[indice]
    return amostra_sistematica
```

```python
amostra_sis = amostra_sistematica(sales_df,5)
amostra_sis.head(1000)
```

#Amostragem sistemática

```python
# Faça uma Funçao para definir a amostragem sistemática
def amostragem_sistematica(df, step):
 
    indexes = np.arange(3, len(df), step = step)
    amostragem_sistematica = df.iloc[indexes]
    return amostragem_sistematica

amostra_sistematica = amostragem_sistematica(vendas, 3)
 
# Visualiza a função
amostra_sistematica.head()
```

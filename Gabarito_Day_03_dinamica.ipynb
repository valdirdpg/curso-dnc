**Plotar gráfico com o valor de vendas em dias normais e em feriados**

```python
df['Date'] =  pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
```

```python
df['month'] = pd.DatetimeIndex(df['Date']).month
```

```python
df.info()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x = 'month',
            y = 'Weekly_Sales',
            hue = 'Holiday_Flag',
            palette = "Blues",
            data = df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import plotly.express as px
import plotly.graph_objects as go
```

```python
house = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
house2 = house.rename(columns={
    'housing_median_age': 'tempo_medio_de_moradia',
    'total_rooms':'quartos_totais',
    'total_bedrooms': 'banheiros_totais',
    'population': 'populacao',
    'households':'familia',
    'median_income': 'renda_media_familiar',
    'median_house_value': 'preco_medio_casa'

})
```

```python
house2
```

```python
house_agrupado = house2.groupby(['tempo_medio_de_moradia']).mean().reset_index()
```

```python
casa_agrupada_filtro = house_agrupado[house_agrupado['tempo_medio_de moradia'] > 20]
```

```python
plt.figure(figsize=(15,8))
plt.title('Preco medio das casas pelo tempo de moradia')
plt.ylabel('Tempo medio na maradia')
plt.xlabel('preco medio das casas')
plt.bar(casa_agrupada_filtro['tempo_medio_de moradia'],casa_agrupada_filtro['preco_medio_casa'], color='red');
plt.show()
```

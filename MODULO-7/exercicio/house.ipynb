```python
labels = np.array(house['SalePrice'])
house= house.drop('SalePrice', axis = 1)
house_list = list(house.columns)
house = np.array(house)
house.shape
```

```python
train_house, test_house, train_labels, test_labals = train_test_split(house, labels, test_size = 0.25, random_state = 42)
```

```python
baseline_preds = test_house[:, house_list.index('OverallQual')]
baseline_errors = abs(baseline_preds - test_labals)
print('Average baseline error: ', round(np.mean(baseline_errors), 2))
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(train_house,train_labels);
```

```python
predictions_rf = rf.predict(test_house)
errors_rf = abs(predictions_rf - test_labals)
r_sq = rf.score(house, labels)
print('Coeficiente de DeterminaÃ§Ã£o (RÂ²):', r_sq)
print('Mean Absolute Error:', round(np.mean(errors_rf), 2), 'degrees.')
print('Mean Squared Error:', metrics.mean_squared_error(test_labals, predictions_rf))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(test_labals, predictions_rf)))
```

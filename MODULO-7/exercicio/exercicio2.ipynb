```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
limite = pd.read_excel('Limite_Credito_Dummy.xlsx')
```

```python
limite.head()
```

```python
limite.shape
```

```python
limites = pd.get_dummies(limite)
```

```python
limites
```

```python
limites.corr()
```

```python
funcao = "LimitedoChequeEspecial~ Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio	+Escolaridade_Superior_Pos	+Gênero_RJ	+Gênero_SP	*Região_F	+Região_M"
modelo = smf.ols(formula=funcao, data=limites).fit()
modelo.summary()
```

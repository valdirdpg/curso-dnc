# SOBRE DROP_DUPLICATE

def drop_duplicates(subset: (Hashable | Sequence[Hashable] | None)=Nenhum, mantenha: (Literal['first'] | Literal['last'] | Literal[False])='first', inplace: bool=False , ignore_index: bool=False) ->(DataFrame | Nenhum)
Retorne DataFrame com linhas duplicadas removidas.

A consideração de determinadas colunas é opcional. Índices, incluindo índices de tempo
são ignorados.

Parâmetros

subconjunto: rótulo da coluna ou sequência de rótulos, opcional
    Considere apenas algumas colunas para identificar duplicatas, por
padrão usa todas as colunas.
manter: {'primeiro', 'último', False}, padrão 'primeiro'
    Determina quais duplicatas (se houver) manter.


first : Elimine as duplicatas, exceto a primeira ocorrência.

last : Solte as duplicatas, exceto a última ocorrência.

Falso: Elimine todas as duplicatas.
inplace: bool, padrão False
Se deve colocar duplicatas no lugar ou devolver uma cópia.
ignore_index : bool, padrão Falso
Se True, o eixo resultante será rotulado como 0, 1, …, n - 1.


devoluções

DataFrame ou nenhum
    DataFrame com duplicatas removidas ou None se inplace=True.

Veja também

DataFrame.value_counts: conta combinações exclusivas de colunas.

Exemplos

Considere o conjunto de dados contendo classificação de ramen.

>>> df = pd.DataFrame({
... 'marca': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
... 'estilo': ['xícara', 'xícara', 'xícara', 'pacote', 'pacote'],
... 'avaliação': [4, 4, 3,5, 15, 5]
... })
>>> df
    classificação de estilo de marca
0 copo Yum Yum 4.0
1 copo Yum Yum 4.0
2 Copa Indomie 3.5
3 Indomie pack 15.0
4 Indomie pack 5.0
Por padrão, ele remove linhas duplicadas com base em todas as colunas.

>>> df.drop_duplicates()
    classificação de estilo de marca
0 copo Yum Yum 4.0
2 Copa Indomie 3.5
3 Indomie pack 15.0
4 Indomie pack 5.0
Para remover duplicatas em colunas específicas, use subconjunto.

>>> df.drop_duplicates(subset=['marca'])
    classificação de estilo de marca
0 copo Yum Yum 4.0
2 Copa Indomie 3.5
Para remover duplicatas e manter as últimas ocorrências, use keep.

>>> df.drop_duplicates(subset=['marca', 'estilo'], manter='último')
    classificação de estilo de marca
1 copo Yum Yum 4.0
2 Copa Indomie 3.5
4 Indomie pack 5.0

## Distribuição Normal Multivariada - Alguns comandos em R

```r
# Variáveis não correlacionadas

mu=c(0,0)
Sigma = matrix(c(2,0,0,1),nrow=2)
```

```r
mu
```

```r
Sigma
```

```r
library(mvtnorm)

x1 <- seq(-5, 5, length= 40)
x2 <- x1

f <- matrix(0, nrow=length(x1), ncol=length(x2))

for (i in 1:length(x1))
  for (j in 1: length(x2))
    f[i,j] <- dmvnorm(c(x1[i],x2[j]), mean=mu, sigma=Sigma)

 persp(x1, x2, f, theta = 70, phi = 30, col = "lightblue", ticktype = "detailed")
```

```r
# Contornos elípticos

contour(x1, x2, f, draw=T, nlevels=20, labcex=0.8, xlab=expression(x[1]),ylab=expression(x[2]))
```

Definindo os parâmetros da distribuição Normal (${\mu}$, ${\Sigma}$),\\ com ${\mu}=\left[\begin{array}{c}1\\0\end{array}\right]$ e ${\Sigma}=\left[\begin{array}{cc}2 & 1\\ 1 & 1\end{array}\right]$:


```r
# Variáveis correlacionadas

mu=c(0,0)
Sigma = matrix(c(2,1,1,1),nrow=2)


library(mvtnorm)

x1 <- seq(-5, 5, length= 100)
x2 <- x1

f <- matrix(0, nrow=length(x1), ncol=length(x2))

for (i in 1:length(x1))
  for (j in 1: length(x2))
    f[i,j] <- dmvnorm(c(x1[i],x2[j]), mean=mu, sigma=Sigma)

 persp(x1, x2, f, theta = 80, phi = 30, col = "lightblue", ticktype = "detailed")
```

```r
contour(x1, x2, f, draw=T, nlevels=20, labcex=0.8, xlab=expression(x[1]),ylab=expression(x[2]))
```

```r
# Simulação de amostras da Normal Multivariada

mu <- c(2,1)
Sigma <- matrix(c(2,1,1,1), nrow=2)

rmvnorm(100, mean=mu, sigma=Sigma);
pairs(rmvnorm(100, mean=mu, sigma=Sigma));
```

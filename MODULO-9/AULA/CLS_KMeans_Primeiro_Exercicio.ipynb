```python
FEATURES = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)'] 
```

```python
df = pd.DataFrame(load_iris()['data'], columns=load_iris()['feature_names'])
df['target'] = load_iris()['target']
dici = {0 : 'setosa', 1: 'versicolor', 2 : 'virginica'} 
df['target_read'] = df['target'].replace(dici)
```

```python
FEATURES = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)'] 
```

```python
for x in FEATURES: 
    plt.figure(figsize= (12,6))
    sns.boxplot(x = 'target_read', y = x, data= df)
    plt.title(x)
    plt.show()
```

```python
sns.pairplot(df, hue= 'target_read')
```

## Implementação

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(3)
```

```python
km.fit(df[FEATURES])
```

```python
km.cluster_centers_
```

```python
km.inertia_
```

```python
km.labels_
```

```python
km.n_iter_
```

```python
km.n_features_in_
```

```python
df['cluster'] = km.predict(df[FEATURES])
```

```python
df
```

```python
df['cluster'] = df.cluster.astype('str')
```

```python
aux = pd.DataFrame(df.groupby('target_read')
                   ['cluster'].value_counts()).rename(columns = {'cluster': 'count'}).reset_index()
```

```python
aux
```

### Análise

```python
for x in FEATURES: 
    plt.figure(figsize= (12,6))
    sns.boxplot(x = 'y_pred', y = x, data= df)
    plt.title(x)
    plt.show()
```

```python
sns.pairplot(df, hue= 'cluster')
```

```python
aux
```

```python
dicionary_class = {'1': 'setosa', '2' : 'versicolor', '0' : 'virginica'}
df['y_pred'] = df['cluster'].replace(dicionary_class)
```

```python
print(classification_report(y_true = df['target_read'], y_pred = df['y_pred']))
```

```python
print(confusion_matrix(y_true = df['target_read'], y_pred = df['y_pred']))
```
